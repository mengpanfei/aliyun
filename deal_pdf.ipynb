```python
import re
import pdfminer
from pdfminer.pdfparser import PDFParser, PDFDocument
from pdfminer.pdfdevice import PDFDevice
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LTTextBoxHorizontal, LAParams
from pdfminer.pdfinterp import PDFTextExtractionNotAllowed

import logging
logging.getLogger("pdfminer").setLevel(logging.ERROR)

pdf=open("data/cecumanalysis5.12.pdf",'rb')
parser = PDFParser(pdf)

doc = PDFDocument()
parser.set_document(doc)
doc.set_parser(parser)
doc.initialize()

# print(doc.is_extractable)

rsrcmagr = PDFResourceManager()
laparams = LAParams()
device = PDFPageAggregator(rsrcmagr, laparams=laparams)
interpreter = PDFPageInterpreter(rsrcmagr, device)

for page in doc.get_pages():
#     print("pageid",page.pageid)
    interpreter.process_page(page)
    layout = device.get_result()
    for x in layout:
        if re.findall("LTRect|LTLine|LTCurve|LTTextLineHorizontal",str(type(x))): break
        for txt in x.get_text().strip().split("\n"):
            if "%" in txt or "[" in txt or "Page" in txt: 
                print(txt)
    print("----------------------")
    for x in layout:
#         print(type(x),dir(x), x.get_text(),x.index,x.height,x.width,x.x0,x.x1,x.y0,x.y1,x)
        if re.findall("LTRect|LTLine|LTCurve|LTTextLineHorizontal",str(type(x))):
            #print(x)
            break
        else:
#             print(type(x),x.index,x.get_text().strip().replace("\n","_"))
#             print(x.index, x.get_text().strip().replace("\n","_"))
            txt=x.get_text().strip()
            if "%" in txt or "[" in txt or "Page" in txt: 
#                 print(txt)
                continue
#             if len(txt)==1 or re.match("\d+",txt): continue
#             print(txt)
        
#     print("----------------------")
```
