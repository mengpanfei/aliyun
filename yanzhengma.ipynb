```python
# opencv 操作图片
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('yzm2.jpg') #读图片
cv2.imwrite("yzm2_.jpg",img) #写图片
plt.subplot(231),plt.imshow(img,'gray'),plt.title('ORIGINAL')
print(img.shape, img.size, img.dtype)
img.copy()

px = img[10,10]
print(px) # 像素点

blue = img.item(10,10,0) #blue = img[10,10,0]
print(blue) #色值

img[10,10] = [255,255,255] #修改像素点
print(img[10,10])

img[10:20,10:20,0] = 254 #修改色值
img.itemset((10,10,1),0) #更高效的方法
print(img[10,10])
plt.subplot(232),plt.imshow(img,'gray'),plt.title('edit')

#复制区域
achar = img[0:20, 0:20]
img[2:22, 70:90] = achar
plt.subplot(233),plt.imshow(img,'gray'),plt.title('copy')

#清除一个通道
img[:,:,2] = 0
plt.subplot(234),plt.imshow(img,'gray'),plt.title('clear channel')

#通道合并操作。 如果用Numpy indexing 会更高效。
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
plt.subplot(235),plt.imshow(img,'gray'),plt.title('clear channel')
img = cv2.merge((r,g,b)) #通道互换
plt.subplot(236),plt.imshow(img,'gray'),plt.title('clear channel')


plt.show()
cv2.imwrite("yzm2_.jpg",img)
```

```python
# 加边框
import cv2
import numpy as np
from matplotlib import pyplot as plt

BLUE = [255,0,0]

img1 = cv2.imread('yzm2.jpg')

replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

plt.show()
```
